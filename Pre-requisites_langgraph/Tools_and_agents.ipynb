### Create Custom Tools

```python
from langchain_core.tools import tool

@tool
def get_word_length(word: str) -> int:
    """Returns the length of the word"""
    return len(word)
```

```python
get_word_length.invoke("Hello")
```

```python
get_word_length.invoke("Geeksforgeeks")
```

```python
@tool
def summation(a:int, b:int) -> int:
    """Returns the sum of two numbers"""
    return a+b
```

```python
print(summation.name)
print(summation.description)
print(summation.args)
```

```python
summation.invoke({'a':1, 'b':2})
```
